```python
#words = ['paper','papers', 'trunk','trunks','pan','pans']

words = ['nach', 'abzüglich','aufgaben','aufgabe']
```

```python
import what3words
import folium
import itertools

import config
```

```python
names =[]
nts = itertools.combinations_with_replacement(words,3)
for nt in nts:
    names.append('.'.join(nt))
```

```python
geocoder = what3words.Geocoder(config.apikey)
res =[]
for name in names:
    
    result = geocoder.convert_to_coordinates(name)
    if 'error' in result.keys():
        print(name + ": not found")
    else:
        res.append(result)
```

```python


startcords = [res[0]['coordinates']['lat'], res[0]['coordinates']['lng']]

# Create the map
m = folium.Map(location=startcords, zoom_start=6, tiles='Stamen Toner')
for r in res:
    try:
        folium.Marker([r['coordinates']['lat'],r['coordinates']['lng']],popup=r['words']).add_to(m)
        print(r['words']+":\t"+ str(r['coordinates'])+"\t"+ r['nearestPlace'])
    except:
        print(r['error']['message'])
```
